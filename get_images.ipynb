```python
#
# I have a folder full of contents from a corrupt ubuntu install (grrrr). This moves through the folder
# and finds all the larger jpeg images, which are probably from gdrive. Move them to a new directory
# for manual review
#

import os
import numpy as np
#from progress.bar import IncrementalBar
from pathlib import Path             #file path tools
import imageio

main_dir = '/home/kgonzalez/REVIEW'
dir_list = os.listdir(main_dir)
print('******')

jpg_list=[]
counter = 0
for subdir in dir_list:
    
    file_list=os.listdir(os.path.join(main_dir, subdir))
    for fname in file_list:
        full_file_name = os.path.join(main_dir,subdir, fname)
        fsize = os.path.getsize(full_file_name)
        if (('.jpg' in fname) and (fsize >= 100000)):
            #store it
            #full_file_name = os.path.join(main_dir,subdir, fname)
            
            im_data =  imageio.imread(full_file_name)
            NN = np.shape(im_data)
            nx = NN[0]
            ny = NN[1]
            del im_data
            if ( (nx >= 900) and (ny >= 900)):
                #print(nx,ny)
                counter = counter + 1
                jpg_list.append(full_file_name) #os.path.join(main_dir,subdir, fname))
                if ((counter % 300) == 0):
                    print('processing file ', counter)

print(len(jpg_list))

```

```python
import pickle

jpg_pickle = '/home/kgonzalez/PYTHON_SCRIPTS/jpg_pickle_list.pck'

with open(jpg_pickle, 'wb') as handle:
    pickle.dump(jpg_list, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
import shutil

out_dir = '/home/kgonzalez/IMAGE_REVIEW'
for file in jpg_list:
    fname = os.path.basename(file)
    if (os.path.exists(file)):
        dst = os.path.join(out_dir,fname)
        if(os.path.exists(dst)): #file in output already, give new name
            dst =  dst + '_copy'

        shutil.copy(file, dst)
```
